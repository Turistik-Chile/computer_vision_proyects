# ReorganizaciÃ³n de Dataset YOLO - Paso a Paso

Este notebook reorganiza el dataset en la estructura:
```
Cumbre_dataset/
â”œâ”€â”€ train/
â”‚   â”œâ”€â”€ images/     (80% de las imÃ¡genes)
â”‚   â””â”€â”€ labels/     (labels correspondientes)
â”œâ”€â”€ test/
â”‚   â”œâ”€â”€ images/     (10% de las imÃ¡genes)
â”‚   â””â”€â”€ labels/     (labels correspondientes)
â”œâ”€â”€ valid/
â”‚   â”œâ”€â”€ images/     (10% de las imÃ¡genes)
â”‚   â””â”€â”€ labels/     (labels correspondientes)
â”œâ”€â”€ images/         (carpeta original)
â”œâ”€â”€ labels/         (carpeta original)
â””â”€â”€ data.yaml
```

**Proceso:**
1. Crear estructura de carpetas (train/test/valid con images y labels)
2. Distribuir imÃ¡genes aleatoriamente (no consecutivas)
3. Mover labels siguiendo a sus imÃ¡genes correspondientes

## ğŸ“¦ Paso 1: Importar librerÃ­as necesarias

````python
import os
import shutil
import random
from pathlib import Path

print("âœ… LibrerÃ­as importadas correctamente")
````

## ğŸ“‚ Paso 2: Definir rutas del dataset

````python
# Ruta base del dataset
dataset_path = Path(r'C:\Users\AlexAravena\Desktop\CV_PROYECT\DATASETS\Oasis_final\Oasis_p2')

# Rutas actuales
images_dir = dataset_path / 'images'
labels_dir = dataset_path / 'labels'

print(f"ğŸ“ Dataset path: {dataset_path}")
print(f"ğŸ–¼ï¸  Images dir: {images_dir}")
print(f"ğŸ·ï¸  Labels dir: {labels_dir}")
print(f"\nÂ¿Existe images? {images_dir.exists()}")
print(f"Â¿Existe labels? {labels_dir.exists()}")
````

## ğŸ—ï¸ Paso 3: Crear estructura de carpetas train/test/valid

````python
# Crear carpetas train, test, valid en la raÃ­z del dataset
splits = ['train', 'test', 'valid']

for split in splits:
    # Crear carpeta principal del split
    split_dir = dataset_path / split
    split_dir.mkdir(parents=True, exist_ok=True)
    print(f"âœ… Creada: {split_dir}")
    
    # Crear subcarpeta images dentro del split
    img_split_dir = split_dir / 'images'
    img_split_dir.mkdir(parents=True, exist_ok=True)
    print(f"   â””â”€â”€ {img_split_dir}")
    
    # Crear subcarpeta labels dentro del split
    lbl_split_dir = split_dir / 'labels'
    lbl_split_dir.mkdir(parents=True, exist_ok=True)
    print(f"   â””â”€â”€ {lbl_split_dir}")

print("\nğŸ‰ Â¡Estructura de carpetas creada exitosamente!")
````

## ğŸ·ï¸ Paso 4: Renombrar imÃ¡genes y labels con prefijo de la carpeta

````python
# Obtener el nombre de la carpeta del dataset (Ãºltima parte del path)
dataset_name = dataset_path.name  # Ejemplo: 'Oasis2', 'Cumbre_dataset', etc.

print(f"ğŸ“ Renombrando archivos con prefijo: '{dataset_name}_'")

# Renombrar todas las imÃ¡genes en la carpeta images
images_renamed = 0
for img_file in images_dir.glob('*.png'):
    # Si ya tiene el prefijo, saltar
    if img_file.name.startswith(f"{dataset_name}_"):
        continue
    
    # Nuevo nombre: dataset_name + nombre original
    new_name = f"{dataset_name}_{img_file.name}"
    new_path = images_dir / new_name
    
    # Renombrar
    img_file.rename(new_path)
    images_renamed += 1

print(f"   âœ… {images_renamed} imÃ¡genes renombradas")

# Renombrar todos los labels en la carpeta labels
labels_renamed = 0
for label_file in labels_dir.glob('*.txt'):
    # Si ya tiene el prefijo, saltar
    if label_file.name.startswith(f"{dataset_name}_"):
        continue
    
    # Nuevo nombre: dataset_name + nombre original
    new_name = f"{dataset_name}_{label_file.name}"
    new_path = labels_dir / new_name
    
    # Renombrar
    label_file.rename(new_path)
    labels_renamed += 1

print(f"   âœ… {labels_renamed} labels renombrados")
print(f"\nğŸ‰ Renombrado completado!")
print(f"   Ejemplo: frame000.png â†’ {dataset_name}_frame000.png")
````

## ğŸ” Paso 5: Listar todas las imÃ¡genes disponibles

````python
# Obtener todas las imÃ¡genes .jpg (solo del directorio raÃ­z de images, no subdirectorios)
all_images = [f for f in images_dir.iterdir() if f.is_file() and f.suffix.lower() == '.png']

print(f"ğŸ“Š Total de imÃ¡genes encontradas: {len(all_images)}")
print(f"\nğŸ”¸ Primeras 5 imÃ¡genes:")
for img in all_images[:5]:
    print(f"  - {img.name}")
````

## ğŸ² Paso 6: Mezclar imÃ¡genes aleatoriamente y dividir en train/test/valid

````python
# Mezclar aleatoriamente con semilla fija (opcional)
import random
random.seed(42)
random.shuffle(all_images)

# Calcular cantidades para cada split
total = len(all_images)
train_size = int(total * 0.8)
test_size = int(total * 0.1)

# Dividir usando slicing simple
train_images = all_images[0:train_size]
test_images = all_images[train_size:train_size + test_size]
valid_images = all_images[train_size + test_size:]

# Mostrar resultados
print(f"ğŸ“Š DivisiÃ³n completada:")
print(f"  ğŸŸ¢ Train: {len(train_images)} imÃ¡genes")
print(f"  ğŸ”´ Test:  {len(test_images)} imÃ¡genes")
print(f"  ğŸŸ¡ Valid: {len(valid_images)} imÃ¡genes")
print(f"  ğŸ“Œ Total: {len(train_images) + len(test_images) + len(valid_images)} imÃ¡genes")
print("\nâœ… Â¡Listo!")
````

## ğŸš€ Paso 7: Mover imÃ¡genes a sus carpetas correspondientes

````python
def mover_imagenes(image_list, split_name):
    """Mueve las imÃ¡genes a la carpeta correspondiente: split/images/"""
    dest_dir = dataset_path / split_name / 'images'
    moved_count = 0
    total = len(image_list)
    
    for i, img_path in enumerate(image_list, 1):
        dest_path = dest_dir / img_path.name
        shutil.move(str(img_path), str(dest_path))
        moved_count += 1
        
        # Mostrar progreso cada 100 imÃ¡genes
        if i % 100 == 0 or i == total:
            print(f"   Progreso: {i}/{total} imÃ¡genes movidas...", end='\r')
    
    print()  # Nueva lÃ­nea al final
    return moved_count

# Mover train
print("ğŸŸ¢ Moviendo imÃ¡genes a TRAIN/images/...")
train_moved = mover_imagenes(train_images, 'train')
print(f"   âœ… {train_moved} imÃ¡genes movidas a train/images/")

# Mover test
print("\nğŸ”´ Moviendo imÃ¡genes a TEST/images/...")
test_moved = mover_imagenes(test_images, 'test')
print(f"   âœ… {test_moved} imÃ¡genes movidas a test/images/")

# Mover valid
print("\nğŸŸ¡ Moviendo imÃ¡genes a VALID/images/...")
valid_moved = mover_imagenes(valid_images, 'valid')
print(f"   âœ… {valid_moved} imÃ¡genes movidas a valid/images/")

print("\nğŸ‰ Â¡Todas las imÃ¡genes han sido distribuidas!")
````

## ğŸ·ï¸ Paso 8: Mover labels correspondientes a train

````python
def mover_labels_para_split(split_name):
    """Mueve los labels que corresponden a las imÃ¡genes de un split especÃ­fico"""
    img_split_dir = dataset_path / split_name / 'images'
    lbl_split_dir = dataset_path / split_name / 'labels'
    
    # Obtener todas las imÃ¡genes en este split
    images_in_split = list(img_split_dir.glob('*.png'))
    total = len(images_in_split)
    
    moved_count = 0
    not_found_count = 0
    
    for i, img_path in enumerate(images_in_split, 1):
        # Obtener el nombre base sin extensiÃ³n
        img_name = img_path.stem  # Ejemplo: 'Cumbre1_0910_frame_0000'
        
        # Buscar el label correspondiente en la carpeta labels original
        label_source = labels_dir / f"{img_name}.txt"
        
        if label_source.exists():
            # Mover el label a la carpeta labels del split
            label_dest = lbl_split_dir / f"{img_name}.txt"
            shutil.move(str(label_source), str(label_dest))
            moved_count += 1
        else:
            not_found_count += 1
            print(f"   âš ï¸  Label no encontrado: {img_name}.txt")
        
        # Mostrar progreso cada 100 archivos
        if i % 100 == 0 or i == total:
            print(f"   Progreso: {i}/{total} labels procesados...", end='\r')
    
    print()  # Nueva lÃ­nea al final
    return moved_count, not_found_count

# Mover labels de TRAIN
print("ğŸŸ¢ Moviendo labels para TRAIN/labels/...")
train_lbl_moved, train_lbl_missing = mover_labels_para_split('train')
print(f"   âœ… {train_lbl_moved} labels movidos a train/labels/")
if train_lbl_missing > 0:
    print(f"   âš ï¸  {train_lbl_missing} labels no encontrados")
````

## ğŸ·ï¸ Paso 9: Mover labels correspondientes a test

````python
# Mover labels de TEST
print("ğŸ”´ Moviendo labels para TEST/labels/...")
test_lbl_moved, test_lbl_missing = mover_labels_para_split('test')
print(f"   âœ… {test_lbl_moved} labels movidos a test/labels/")
if test_lbl_missing > 0:
    print(f"   âš ï¸  {test_lbl_missing} labels no encontrados")
````

## ğŸ·ï¸ Paso 10: Mover labels correspondientes a valid

````python
# Mover labels de VALID
print("ğŸŸ¡ Moviendo labels para VALID/labels/...")
valid_lbl_moved, valid_lbl_missing = mover_labels_para_split('valid')
print(f"   âœ… {valid_lbl_moved} labels movidos a valid/labels/")
if valid_lbl_missing > 0:
    print(f"   âš ï¸  {valid_lbl_missing} labels no encontrados")

print("\nğŸ‰ Â¡Todos los labels han sido distribuidos!")
````

## ğŸ“Š Paso 11: Verificar la distribuciÃ³n final

````python
# Contar archivos en cada carpeta
for split in ['train', 'test', 'valid']:
    img_count = len(list((dataset_path / split / 'images').glob('*.png')))
    lbl_count = len(list((dataset_path / split / 'labels').glob('*.txt')))
    
    print(f"\nğŸ“ {split.upper()}:")
    print(f"   ğŸ–¼ï¸  ImÃ¡genes: {img_count}")
    print(f"   ğŸ·ï¸  Labels:   {lbl_count}")
    
    if img_count == lbl_count:
        print(f"   âœ… Â¡Correspondencia perfecta!")
    else:
        print(f"   âš ï¸  Hay diferencia de {abs(img_count - lbl_count)} archivos")
````

## ğŸ“ Paso 12: Preparar data.yaml para Google Colab

````python
import yaml

# Leer el data.yaml actual
data_yaml_path = dataset_path / 'data.yaml'

# Si existe, leer el contenido actual
if data_yaml_path.exists():
    with open(data_yaml_path, 'r') as f:
        data = yaml.safe_load(f)
    print("ğŸ“„ Contenido actual de data.yaml:")
    print(data)
else:
    # Crear estructura bÃ¡sica si no existe
    data = {
        'names': {0: 'person'},  # Ajusta segÃºn tus clases
        'nc': 1  # NÃºmero de clases
    }
    print("ğŸ“„ Creando nuevo data.yaml...")

# ConfiguraciÃ³n para Google Colab
# La ruta 'path' apuntarÃ¡ a donde subes el dataset en Colab (ej: /content/dataset)
dataset_name = dataset_path.name

# Rutas relativas desde el path base (funcionan tanto local como en Colab)
data['path'] = f'/content/{dataset_name}'  # Ruta tÃ­pica en Google Colab
data['train'] = 'train/images'
data['val'] = 'valid/images'
data['test'] = 'test/images'

# Guardar el archivo actualizado
with open(data_yaml_path, 'w') as f:
    yaml.dump(data, f, default_flow_style=False, sort_keys=False)

print("\nâœ… data.yaml actualizado para Google Colab:")
print(f"   ğŸ“ Path en Colab: {data['path']}")
print(f"   ğŸŸ¢ Train: {data['train']}")
print(f"   ğŸŸ¡ Val: {data['val']}")
print(f"   ğŸ”´ Test: {data['test']}")

print("\nğŸ“ Contenido completo del data.yaml:")
print(yaml.dump(data, default_flow_style=False, sort_keys=False))

print("\nğŸ’¡ Instrucciones para Google Colab:")
print(f"   1. Sube la carpeta '{dataset_name}' a /content/ en Colab")
print(f"   2. O usa: !unzip tu_dataset.zip -d /content/")
print(f"   3. La ruta serÃ¡: /content/{dataset_name}/")
print("\nğŸ‰ Â¡Proceso completado exitosamente!")
````
